**Обработка данных**

```python
df.isna().sum() #считает кол-во пропусков по столбцам
```

```python
df = df.fillna('N/A') #заполняет пустоту значениями N/A
df.isna().sum() #проверяем сумму пустых клеток, количество их должно равняться 0
```

```python
column_without_PersonID = [] #Заводим список который будем заполнять
for name_of_column in df.columns: #проходимся по столбцам
    if name_of_column != 'Person ID':
        column_without_PersonID.append(name_of_column) #добавляем в список все столбцы кроме Person ID


number_dubl = df.duplicated(subset=column_without_PersonID).sum() #Проверяет дубликаты у column_without_PersonID

print(f"Количество дубликатов данных (без учета Person ID): {number_dubl}")
```

```python
for i in df.columns:
    print(f"{i}: {df[i].unique()}")# проходимся по столбцам и находим уникальные значения чтобы посмотреть уникальные значения которые присваиваются
```

```python
#замечаем названия имеющие один и тот же смысл 'Normal' и 'Normal Weight'
df['BMI Category'] = df['BMI Category'].str.replace('Normal Weight', 'Normal') #заменяем Normal Weight на Normal
df['BMI Category'].unique()  #выводим также уникальные элементы но уже без Normal Weight
```

**Визуализация данных**

```python
# 1. Гистограмма (распределение возраста)
plt.figure(figsize=(10,6))           # создаем лист для графика размером 10x6 дюймов
sns.histplot(df['Age'], bins=30)     # строим гистограмму для столбца 'Age' с 30 интервалами
plt.ylabel('Количество')             # добавляем описание вертикальной прямой
plt.xlabel('Возраст')                # добавляем описание горизонтальной прямой
plt.title('Распределение возраста')  # добавляем заголовок
plt.grid(True, alpha=0.3)            # делаем сетку на графике  # добавляем заголовок
plt.show()                           # выводим график
```

```python
# 2. Линейный график (средний пульс по возрастам)
plt.figure(figsize=(10,6))                    # создаем лист для графика размером 10x6 дюймов
df.groupby('Age')['Heart Rate'].mean().plot() # группируем людей по возрасту, берем значения пульса, находим среднее арифмитическое, рисуем линейный график
plt.ylabel('Средний пульс (уд/мин)')          # добавляем описание вертикальной прямой
plt.xlabel('Возраст')                         # добавляем описание горизонтальной прямой
plt.title('Средний пульс по возрастам')       # добавляем заголовок
plt.grid(True, alpha=0.3)                     #делаем сетку на графике
plt.show()                                    # выводим график
```

```python
# 3. столбчатые диаграммы (количество людей по профессиям)
plt.figure(figsize=(10,6))                       # создаем лист для графика размером 10x6 дюймов
df['Occupation'].value_counts().plot(kind='bar') # берем столбец Occupation, считаем количество и заносим в диаграмму
plt.xticks(rotation=30, ha='right')              # чтобы не накладывались друг на друга названия профессий, поворачиваем на 45 градусов и отступаем справа
plt.tight_layout()                               # делает достаточно отступов чтобы ничего не накладывалось друг на друга
plt.ylabel('Count')                              # добавляем описание вертикальной прямой
plt.title('Количество людей по профессиям')      # добавляем заголовок
plt.grid(True, alpha=0.3)                        # делаем сетку на графике
plt.show()                                       # выводим график
```

```python
# 4. bloxplot (качество сна по полу)
plt.figure(figsize=(10, 6))                             # создаем лист для графика размером 10x6 дюймов
sns.boxplot(data=df, x='Gender', y='Quality of Sleep')  # создаем bloxplot, берем данные из нашего df и указываваем название столбцов по осям диаграммы
plt.title('Качество сна по полу')                       # добавляем заголовок
plt.xlabel('Пол')                                       # добавляем описание горизонтальной прямой
plt.ylabel('Качество сна (1-10 ч)')                     # добавляем описание вертикальной прямой
plt.grid(True, alpha=0.3)                               # делаем сетку на графике
plt.show()                                              # выводим график
```

```python

plt.figure(figsize=(10,6))                                     # создаем лист для графика размером 10x6 дюймов
# Считаем корреляции между числовыми колонками
Corr_Matrix = df[['Age', 'Sleep Duration', 'Quality of Sleep',
                   'Stress Level', 'Heart Rate']].corr()       # берем столбцы датасета и получаем матрицу 5х5
sns.heatmap(Corr_Matrix , annot=True, cmap='coolwarm')          #создает heatmap корреляций где выставлены числовые значения и выведены цвета
plt.title('Корреляции между показателями')                     # добавляем заголовок
plt.show()                                                     # выводим график
```

**Группировки и агрегации**

```python
# 1. Средняя продолжительность сна по профессиям:
Avg_Sleep_By_Occupation = df.groupby('Occupation')['Sleep Duration'].mean() # группируем профессии по сну и ищем среднее арифметическое
print(Avg_Sleep_By_Occupation.sort_values(ascending=False))                 # выводим значения по убыванию
```

```python
Gender_DailySteps_Sum = df.groupby('Gender')['Daily Steps'].sum() # сумма количества шагов в день по полу
print(Gender_DailySteps_Sum)
```

```python
count_by_gender = df.groupby('Gender').size() # подсчет кол-во людей по полу
print(count_by_gender)
```
